```python
from cenpy import products
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dc = products.ACS(2017).from_place('Washington, DC',
                                   level='tract',
                                   variables=['B00002*', 'B01002H_001E'])
```

```python
f, ax = plt.subplots(figsize=(20, 20))
dc.dropna(subset=['B00002_001E'], axis=0).plot('B00002_001E',
                                               ax=ax,
                                               cmap='plasma')
ax.set_facecolor('k')
```
